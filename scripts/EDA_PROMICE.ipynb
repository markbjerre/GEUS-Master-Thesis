```python
import pandas as pd
import numpy as np

# Load your data into a pandas DataFrame (replace "data.csv" with the name of your file)


# Define the date column that you want to group by (replace "date_column" with the name of your column)
min_values_per_day = 20

# Group the data by weather station and date
grouped = df.groupby(['stid','date'])

# Calculate the number of non-NaN values for each variable within each group
non_nan_counts = grouped.apply(lambda x: x.notnull().sum())

# Filter out groups where any variable has less than min_values_per_day non-NaN values
filtered_groups = grouped.filter(lambda x: (x.notnull().sum() >= min_values_per_day).all())

# Get a list of the columns to average (i.e. all columns except for the date column)
columns_to_average = df.select_dtypes(exclude=['object']).columns

# Group the data by date and calculate the mean for each column
# Calculate the mean for each variable within each group
means = filtered_groups.groupby(['stid','date'])[columns_to_average].mean()

# Display the results
print(means)
```
